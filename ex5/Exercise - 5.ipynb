##### Read the dataset.

```python
import pandas as pd
data = pd.read_csv('phone_number.csv')
```

```python
data
```

```python
import re
# Define a regex pattern
pattern = r'^[7-9][0-9]{9}$'

for i in data['PhoneNumber']:
# Use re.findall to find all matches in the string
    matches = re.findall(pattern, i)

    # Print the result
    if(matches!=[]):
        print(matches)
```
